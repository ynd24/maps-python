```python
lat = []
long = []
```

```python
Coord = []
```

```python
for c in coord:
    x = re.split(",", c)
    Coord.append(x)

```

```python
lat3 = []
long3 = []
for lat_long in Coord:
    lat3.append(float(lat_long[0]))
    long3.append(float(lat_long[1]))
```

```python
coordinate_2 = Coord.copy()
```

```python
for sublist in coordinate_2:
    for element in sublist:
        if sublist.index(element) == 1:
            long.append(element)
        else:
            lat.append(element)
```
